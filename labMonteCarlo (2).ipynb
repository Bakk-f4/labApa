```python
import random
import copy
import math
```

```python
nodi = 9
grafo = [["0", "7"], ["0", "1"], ["0", "2"], ["0", "3"], ["0", "8"],
    ["1", "2"], ["1", "4"], ["1", "7"], ["2", "3"], ["2", "5"], ["2", "4"],
    ["3", "5"], ["3", "8"], ["4", "5"], ["4", "6"], ["4", "7"],
    ["5", "6"], ["5", "8"], ["6", "7"], ["6", "8"], ["7", "8"]]
```

```python
def diffEdge(tupla1, tupla2):
    if tupla1[0] == tupla2[0] and tupla1[1] == tupla2[1]:
        return False
    if  tupla2[0] == tupla1[0] and tupla2[1] == tupla1[1]:
        return False
    else:
        return True   
```

```python
def minCut(g):
    for j in range(nodi, 2, -1):
        arco = random.choice(g)
        
        listaArchi = []
        for elem in g:
            if diffEdge(elem, arco):
                listaArchi.append(elem)
        
        g = listaArchi
       
        n1 = arco[0]
        n2 = arco[1]
        newNode = n1 + n2

 
        for elem in g:
            if n1 in elem:
                elem.remove(n1)
                elem.append(newNode)      
            if n2 in elem:
                elem.remove(n2)
                elem.append(newNode)
                
    #print(g)
    return len(g)
```

```python
tot = 0
giri = 10**5

for i in range(giri):
    copyGrafo = copy.deepcopy(grafo)
    if minCut(copyGrafo) == 4: 
        tot+=1
        
varEmp = tot/giri
print(varEmp)
#
#otteniamo una varianza empirica maggiore a 2/n^2, dove 'n' e' il numero dei nodi
#
```
